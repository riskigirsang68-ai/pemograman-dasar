```python
# Fungsi untuk menghitung efisiensi turbin
def hitung_efisiensi(input_energi, output_energi):
    # Pastikan input_energi tidak nol untuk menghindari pembagian nol
    if input_energi == 0:
        return "Error: input_energi tidak boleh nol."

    efisiensi = output_energi / input_energi
    return efisiensi

# Contoh pemanggilan fungsi (bonus)
input_energi = 500
output_energi = 400

hasil = hitung_efisiensi(input_energi, output_energi)
print(f"Efisiensi turbin = {hasil:.2f} atau {hasil*100:.2f}%")
```

```python
def hitung_efisiensi_termal(energi_keluar, energi_masuk):
    efisiensi = (energi_keluar / energi_masuk) * 100
    print("Efisiensi termal =", efisiensi, "%")

# Contoh pemanggilan fungsi
hitung_efisiensi_termal(400, 500)
```

```python
# Simulasi pengisian tangki bahan bakar
tangki = 0

while tangki <= 100:
    print("Volume tangki:", tangki, "liter")

    if tangki == 80:
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
        break

    tangki += 10
```
